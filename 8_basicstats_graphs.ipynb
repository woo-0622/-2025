```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)

# Generate VOT data for fortis stops (short VOT)
fortis_vot = np.random.normal(loc=15, scale=5, size=30)  # mean = 15ms, std = 5ms

# Generate VOT data for lenis stops (intermediate VOT)
lenis_vot = np.random.normal(loc=50, scale=10, size=30)  # mean = 50ms, std = 10ms

# Generate VOT data for aspirated stops (long VOT)
aspirated_vot = np.random.normal(loc=80, scale=10, size=30)  # mean = 80ms, std = 10ms

# Create a DataFrame
data = pd.DataFrame({
    'VOT': np.concatenate([fortis_vot, lenis_vot, aspirated_vot]),
    'Stop Type': ['Fortis']*30 + ['Lenis']*30 + ['Aspirated']*30
})
# Save the data to a CSV file (optional)
data.to_csv('korean_stops_vot.csv', index=False)

# Print the first few rows of the data
print(data.head())

# Plot the data using seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(x='Stop Type', y='VOT', data=data, palette='pastel')
plt.title('VOT Distribution for Korean Stops')
plt.xlabel('Stop Type')
plt.ylabel('VOT (ms)')
plt.show()
```
