```python
#@markdown  **ğŸ¹ Student Activity 8 of the previous class material-Extended2 â¤µï¸**

#@markdown  **Choose rows í–‰ ì„ íƒ**

#@markdown  **Index operator[ ] with integer location of the rows**

import pandas as pd

exam_data = {'Math':[90, 80, 70, 85], 'English':[98, 89, 95, 79],
             'Music':[85, 95, 100, 100], 'Gym':[100,90,90, 88]}

df = pd.DataFrame(exam_data, index = ['Simon', 'Emma', 'Mary', 'Louis'])
print(df)
print('\n')

df1 = df.iloc[::2]
print(df1)
print('\n')

df2 = df.iloc[0:3]
print(df2)
print('\n')

df3 = df.iloc[::-1]
print(df3)
```

```python
#-*- coding: utf-8 -*-

import pandas as pd

#DataFrame( ) í•¨ìˆ˜ë¡œ ì‚¬ì „í˜• ë°ì´í„°ë¥¼ 2ì°¨ì› ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜, ë³€ìˆ˜ dfì— ì €ì¥.
#ì£¼ì˜: í•™ìƒë“¤ ì´ë¦„ì´ ìˆëŠ” Name column & ê° ê³¼ëª© ì„±ì  columns ì„ì—¬ ìˆìŒ.

exam_data = {'Name' : ['Simon', 'Mathew', 'Andy','Louis'],
             'Math':[90,80,70,85],
             'English':[98,89,95,79],
             'Music':[85,95,100,100],
             'Physics':[100,90,90,88]}

df = pd.DataFrame(exam_data)
print(df)
print('\n')

#'Name' columnì„ set_index() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ìƒˆë¡œìš´ ì¸ë±ìŠ¤ë¡œ ì§€ì •í•˜ê³ , df objectì— ë³€ê²½ ì‚¬í•­ ë°˜ì˜; inplace=True means an immediate application without giving a new variable name.
df.set_index('Name', inplace=True)
print(df)

```

####<font color = 'red'> ğŸ‘€ğŸ¾ **ë°ì´í„°í”„ë ˆì„ ë³€í™˜ ì—°ìŠµ ê³„ì† Converting to basic data frame from dictionary data** 1 â¤µï¸

```python
#-*- coding: utf-8 -*-

import pandas as pd

#DataFrame( ) í•¨ìˆ˜ë¡œ ë°ì´í„°í”„ë ˆì„ ë³€í™˜, ë³€ìˆ˜ dfì— ì €ì¥

dict_data = {'c0' : [1,2,3],
             'c1' : [4,5,6],
             'c2' : [7,8,9],
             'c3' : [10,11,12],
             'c4' : [13,14,15]}

df = pd.DataFrame(dict_data)
print(df)
print('\n')
```

```python
df= pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(df)
print('\n')
```

```python
#Dictionaryë°ì´í„°ë¥¼ í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•˜ë©°, í–‰ ì¸ë±ìŠ¤ë¥¼ 'r0', 'r1', 'r2'ë¡œ ì¬ì§€ì •

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df)
print('\n')
```
